```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def run_eda():

    BASE_DIR = os.path.abspath("..")
    output_dir = os.path.join(BASE_DIR, "plots")
    os.makedirs(output_dir, exist_ok=True)

    # ✅ Load TRAIN + TEST (IMPORTANT FIX)
    train = pd.read_csv("../data/processed/train.csv")
    test = pd.read_csv("../data/processed/test.csv")

    df = pd.concat([train, test], axis=0)

    # Date conversion
    df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month

    # ==============================
    # 1. YEARLY SALES
    # ==============================
    yearly_sales = df.groupby('Year')['Weekly_Sales'].sum() / 1e6

    plt.figure(figsize=(8,6))
    yearly_sales.plot(kind='bar')
    plt.title("Total Sales by Year")
    plt.ylabel("Sales (Millions)")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "sales_by_year.png"))
    plt.close()

    # ==============================
    # 2. STORE AVG SALES
    # ==============================
    plt.figure(figsize=(12,6))
    sns.barplot(x='Store', y='Weekly_Sales', data=df, estimator='mean', ci=None)
    plt.title("Average Sales per Store")
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "store_avg_sales.png"))
    plt.close()

    # ==============================
    # 3. HOLIDAY IMPACT
    # ==============================
    plt.figure()
    sns.barplot(x='Holiday_Flag', y='Weekly_Sales', data=df)
    plt.title("Holiday vs Sales")
    plt.savefig(os.path.join(output_dir, "holiday_sales.png"))
    plt.close()

    # ==============================
    # 4. CORRELATION
    # ==============================
    plt.figure(figsize=(10,8))
    corr = df[['Weekly_Sales','Temperature','Fuel_Price','CPI','Unemployment']].corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Correlation Heatmap")
    plt.savefig(os.path.join(output_dir, "correlation.png"))
    plt.close()

    # ==============================
    # 5. MONTHLY TREND (YEAR-WISE)
    # ==============================
    for year in df['Year'].unique():
        yearly_data = df[df['Year'] == year]
        monthly_sales = yearly_data.groupby('Month')['Weekly_Sales'].sum() / 1e6

        plt.figure()
        monthly_sales.plot(marker='o')
        plt.title(f"Monthly Trend - {year}")
        plt.savefig(os.path.join(output_dir, f"monthly_{year}.png"))
        plt.close()

    # ==============================
    # 6. TOP STORES
    # ==============================
    plt.figure()
    top_stores = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(10)
    top_stores.plot(kind="bar")
    plt.title("Top 10 Stores")
    plt.savefig(os.path.join(output_dir, "top_stores.png"))
    plt.close()

    # ==============================
    # 7–10 SCATTER PLOTS
    # ==============================
    features = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

    for feature in features:
        plt.figure()
        sns.scatterplot(x=feature, y='Weekly_Sales', data=df)
        plt.title(f"{feature} vs Sales")
        plt.savefig(os.path.join(output_dir, f"{feature}_vs_sales.png"))
        plt.close()

    # ==============================
    # 11. WEEKLY PATTERN
    # ==============================
    df['Day_of_week'] = df['Date'].dt.dayofweek

    plt.figure()
    sns.boxplot(x='Day_of_week', y='Weekly_Sales', data=df)
    plt.title("Sales by Day of Week")
    plt.savefig(os.path.join(output_dir, "weekly_pattern.png"))
    plt.close()

    print("✅ EDA completed. Plots saved at:", output_dir)


if __name__ == "__main__":
    run_eda()
```
